```python
import requests
from bs4 import BeautifulSoup

headers = {
    'User-Agent': 'Seu nome, example.com',
    'From': 'email@example.com',
}

url = 'https://www.billboard.com/charts/hot-100'

page = requests.get(url, headers=headers)
soup = BeautifulSoup(page.text, 'html.parser')
```

```python
# NA PÁGINA WEB, ENCONTRA OS COMPONENTES QUE POSSUEM ESSA CLASSE CSS
lista_musicas = soup.find_all(class_='o-chart-results-list-row-container')
```

```python
print(f'Número de músicas encontradas: {len(lista_musicas)}')
```

```python
for item in lista_musicas:
    print(item)
```

```python
lista_musicas_itens = []
for musica in lista_musicas:
    extract = musica.find(class_='c-title')
    lista_musicas_itens.append(extract)
```

```python
print(f'Número de músicas encontradas: {len(lista_musicas_itens)}')
```

```python
for item in lista_musicas_itens:
    print(item.prettify())
```

```python
print(lista_musicas_itens[0])
```

```python
nomes_musicas = []

for item in lista_musicas_itens:
    # Supondo que o título esteja na primeira posição da lista de conteúdos
    titulo = item.contents[0] if item.contents else None  
    if titulo is not None:  # Verifique se o título não é None
        # Remover tabulações e quebras de linha, e converter para string
        titulo_limpo = str(titulo).replace('\t', '').replace('\n', '').strip()  
        nomes_musicas.append(titulo_limpo)
        # print(titulo_limpo)
```

```python
print(f'Número de músicas encontradas: {len(nomes_musicas)}')
```

```python
print(nomes_musicas)
```

```python
nome_busca = "b"
musica_encontrada = False  # Para verificar se a música foi encontrada

for nome in nomes_musicas:
    if nome_busca.lower() in nome.lower():  # Usando o operador 'in'
        print("ENCONTREI A MÚSICA...: ", nome)
        musica_encontrada = True  # Marca que a música foi encontrada

if not musica_encontrada:
    print("Música não encontrada.")
```
