```python
# Função para extrair densidade demográfica
async def extrairDensidadeDemografica(url):
    print('Extraindo Densidade de:', url)
    
    async with async_playwright() as p:
        browser = await p.chromium.launch()
        page = await browser.new_page()
        
        # Navegar para a URL
        await page.goto(url)

        # Aguardar um pouco para o conteúdo ser carregado
        await page.wait_for_timeout(300)  # Aguarda 2 segundos (ajuste conforme necessário)

        # Extrair o texto da página
        full_text = await page.content()

        # Procurar pela densidade demográfica
        match = re.search(r'densidade demográfica\s+era\s+de\s+([\d,.]+)', full_text, re.IGNORECASE)
        
        if match:
            densidade = match.group(1)  # Captura o valor numérico
            print("Densidade demográfica encontrada:", densidade)
            await browser.close()
            return densidade
        else:
            print('Densidade demográfica não encontrada no texto.')
            await browser.close()
            return None
```

```python
# Função para executar a extração
async def main():
    inicio = datetime.now()
    inicio_formatado = inicio.strftime("%Y-%m-%d %H:%M:%S")
    print("Data/Hora Início:", inicio_formatado)
    
    dic_municipio_densidade = {}
    for i, municipio in enumerate(municipios):  # Usando a lista municipios
        url = 'https://cidades.ibge.gov.br/brasil/' + estado + '/' + municipios_url[i] + '/panorama'
        densidadeDemografica = await extrairDensidadeDemografica(url)
        print("Densidade demográfica de", municipio, ":", densidadeDemografica)
        dic_municipio_densidade[municipio] = densidadeDemografica

    # Exibir o dicionário resultante
    print(dic_municipio_densidade)

    # Criar novos dicionários baseados na densidade
    municipios_maior_110 = {municipio: densidade for municipio, densidade in dic_municipio_densidade.items() if densidade is not None and float(densidade.replace('.', '').replace(',', '.')) > 110}
    municipios_entre_100_e_110 = {municipio: densidade for municipio, densidade in dic_municipio_densidade.items() if densidade is not None and 100 < float(densidade.replace('.', '').replace(',', '.')) <= 110}
    municipios_menor_100 = {municipio: densidade for municipio, densidade in dic_municipio_densidade.items() if densidade is not None and float(densidade.replace('.', '').replace(',', '.')) < 100}

    # Exibir os novos dicionários
    print("Municípios com densidade maior que 110:", municipios_maior_110)
    print("Municípios com densidade entre 100 e 110:", municipios_entre_100_e_110)
    print("Municípios com densidade menor que 100:", municipios_menor_100)

    fim = datetime.now()
    fim_formatado = fim.strftime("%Y-%m-%d %H:%M:%S")
    print("Data/Hora Fim:", fim_formatado)

# Executar a função principal
asyncio.run(main())
```
