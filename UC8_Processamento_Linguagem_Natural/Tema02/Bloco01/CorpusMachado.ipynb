```python
noticiasTruncado = noticias[:100000]
noticias_limpo = []  # Inicializa uma nova lista para armazenar os tokens limpos

for token in noticiasTruncado:
    if token not in stopwords.words('portuguese'):
        noticias_limpo.append(token)  # Adiciona o token à nova lista se não for uma stopword
```

```python
print(noticias_limpo)
```

```python
print(len(noticiasTruncado))
print(len(noticias_limpo))
```

```python
from nltk.probability import FreqDist
freq = FreqDist(noticias_limpo)
print("Palavras mais frequentes (após limpeza)")
freq.most_common(20)
```

```python
sinais = [',','"','(',')','%']
noticiasTruncado = noticias[:100000]
noticias_limpo = []  # Inicializa uma nova lista para armazenar os tokens limpos

for token in noticiasTruncado:
    if token not in stopwords.words('portuguese') and token not in sinais:
        noticias_limpo.append(token)  # Adiciona o token à nova lista se não for uma stopword
```

```python
print(len(noticiasTruncado))
print(len(noticias_limpo))
```

```python
# from nltk.probability import FreqDist
freq = FreqDist(noticias_limpo)
print("Palavras mais frequentes (após segunda limpeza)")
freq.most_common(20)
```
