```python
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from datetime import datetime
nltk.download('wordnet')
```

```python
print("Data/Hora Início: ", datetime.now())

# url_youtube = "https://www.youtube.com/watch?v=KJzBxJ8ExRk" # GAROTA DE IPANEMA
url_youtube = "https://www.youtube.com/watch?v=1NDxPTLxcrg" # ASTRONAUTA DE MÁRMORE

# Especifique o caminho do executável do ChromeDriver
service = Service(executable_path='/Users/valter/Documents/dev/Estudos/Pos-IA-ML/UC8_Processamento_Linguagem_Natural/Tema04/Bloco02/chromedriver-mac-arm64/chromedriver')

# Configurar as opções do Chrome (opcional, mas recomendado para evitar problemas)
options = webdriver.ChromeOptions()
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')
options.add_argument('--disable-gpu')
options.add_argument('--headless')  # Opcional: executa o Chrome sem interface gráfica

# Inicie o driver com o serviço e opções
with webdriver.Chrome(service=service, options=options) as driver:
    wait = WebDriverWait(driver, 15)
    driver.get(url_youtube)

    # Role a página para carregar mais comentários
    for item in range(30):
        wait.until(EC.visibility_of_element_located((By.TAG_NAME, "body"))).send_keys(Keys.END)
        time.sleep(3)  # Ajuste o tempo de espera para acelerar ou desacelerar a rolagem

    # Coleta os comentários visíveis
    data = []
    comments = wait.until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, "#content #content-text")))
    for comment in comments:
        data.append(comment.text)

# Criar DataFrame com os comentários
df = pd.DataFrame(data, columns=['comment'])
print(df.head())

print("Data/Hora Fim...: ", datetime.now())
```

```python
import re, string
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from textblob import TextBlob
```

```python
sw = stopwords.words('portuguese')

def clean_text(text):
    # print('text....: ', text)
    text = text.lower()
    text = re.sub('@', '', text)
    text = re.sub(r'\[.*?\]', '', text)
    text = re.sub(r'https?://\S+|www\.\S+', '', text)
    text = re.sub('<.*?>+', '', text)
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub('\n', '', text)
    text = re.sub(r'\w*\d\w*', '', text)
    text = re.sub('[^a-zA-Z]+', ' ', text)

    # Tokenize the data
    text = nltk.word_tokenize(text)
    # Remove stopwords
    text = [w for w in text if w not in sw]

    print('text....: ', text)
    
    return text
```

```python
df['comment_manipulacao'] = df['comment'].apply(lambda x: clean_text(x))

lemmatizer = WordNetLemmatizer()

def lem(text):
    text = [lemmatizer.lemmatize(t) for t in text]
    text = [lemmatizer.lemmatize(t, 'v') for t in text]

    print('text 2....: ', text)
    
    return text

df['comment_manipulacao'] = df['comment_manipulacao'].apply(lambda x: lem(x))
```

```python
# GRÁFICO LISTA COM TODAS AS PALAVRAS
all_words = []

for i in range(len(df)):
    all_words = all_words + df['comment_manipulacao'][i]
nlp_words = nltk.FreqDist(all_words)
plot1 = nlp_words.plot(30, color='salmon', title='Frequência de palavras')
```

```python
bigrm = list(nltk.bigrams(all_words))
words_2 = nltk.FreqDist(bigrm)
words_2.plot(30, color='salmon', title='Frequência de palavras')
```
