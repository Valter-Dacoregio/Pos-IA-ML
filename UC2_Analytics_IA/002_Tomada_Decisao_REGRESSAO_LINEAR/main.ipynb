```python
# Importaes de bibliotecas necessrias para anlise e visualizao de dados, alm de aprendizado de mquina
import matplotlib.pyplot as plt # Para a criao de grficos
import numpy as np # Biblioteca para operaes matemticas complexas e manipulao de arrays
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score # Para calcular mtricas de avaliao do modelo

# Dados fornececidos pelo usurio, onde os gastos com publicidade esto em milhares de dlares e as vendas em unidades
gastos_publicidade = [1.5, 2.0, 2.5, 3.0, 3.5, 4.0, 4.5, 5.0, 5.5, 6.0, 6.5, 7.0]
licencas_vendidas =  [ 50,  55,  60,  68,  72,  80,  85,  88,  90,  95, 100, 105]

# Converso das listas em arrays numpy, formato necessrio para uso scikit-learn, e ajuste para valores em reais
X = np.array(gastos_publicidade).reshape(-1, 1) * 1000 # Gastos com publicidade em Real
y = np.array(licencas_vendidas) # Nmero de licenas vendidas

# Criao do modelo de regresso linear utilizando a biblioteca scikit-learn
modelo = LinearRegression()
modelo.fit(X, y) # Treinamento do modelo

y_pred = modelo.predict(X) # Predio dos valores de y com base nos valores de X.
mse = mean_squared_error(y, y_pred) # Clculo do erro quadrtico mdio
r2 = r2_score(y, y_pred) # Clculo do erro quadrtico mdio

# Gerao e configurao do grfico para visualizar os dados a linha de regresso
plt.figure(figsize=(10, 6)) # Definio do tamanho da figura do grfico
plt.scatter(X, y, color='blue', label='Dados reais') # Grfico de disperso para os dados reais
plt.plot(X, y_pred, color='red', linewidth=2, label='Linha de regresso') # Linha de regresso

# Adio de ttulos e rtulos aos eixos do grfico para melhor compreenso
plt.title('Regresso Linear - Vendas de Licenas vs. Gastos com Publicidade')
plt.xlabel('Gastos com Publicidade (R$)')
plt.ylabel('Licenas Vendidas')
plt.legend() # Incluso de legenda no grfico

plt.show() # Exibio do grfico

# Uso do modelo treinado para prever o nmero de licenas vendidas com um gasto de $8.000 em publicidade
previsao = modelo.predict(np.array([[8000]]))

# Os resultado das mtricas de avaliao e a previso so apresentados
# mse, r2, previsao
print("mse.....: ", mse)
print("r2......: ", r2)
print("previsao: ", previsao)
```
